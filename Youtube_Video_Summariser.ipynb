```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_video_id(url):
  video_id = url.split("=")[-1]
  return video_id
```

```python
videoid = get_video_id("https://www.youtube.com/watch?v=xDQL3vWwcp0")
```

```python
transcript = YouTubeTranscriptApi.get_transcript(videoid)
```

```python
transcript
```

```python
transcript_joined = " ".join([i['text'] for i in transcript])
transcript_joined
```

```python
import re
transcript_processed = transcript_joined.lower()
transcript_processed = re.sub(r'[^\w\s]', '', transcript_processed)

transcript_processed
```

```python
import google.generativeai as genai

api_key = "YOUR_API_KEY"

genai.configure(api_key=api_key)

question = "Who is she"

prompt = f"""Answer the question based on the following transcript:

"{transcript_processed}"

Question: {question}
"""

model = genai.GenerativeModel('gemini-2.0-flash')
response = model.generate_content(prompt)

print(response.text)
```
